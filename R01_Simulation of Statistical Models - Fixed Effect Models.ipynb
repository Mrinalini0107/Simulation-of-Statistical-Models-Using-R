## 📘 Simulation of Statistical Models - Fixed Effect Models

## 📌 Fixed Effect Model (Simple t-test with a Two-Level Factor):

### 🔹 Step 1: Define the Factor (Independent Variable)

```r
# Create a two-level factor: Gender
# 2 levels (Male and Female) with 15 replications each
Gender <- gl(n = 2, k = 15, labels = c("Male", "Female"))

# Check the structure of the factor
str(Gender)
```

### 🔹 Step 2: Define the Response Variable (Expected Means)

```r
# Assign expected values:
# Male -> 1.85
# Female -> 1.89

Gender_N <- rep(c(1.85, 1.89), each = 15)

# Check structure of response variable
str(Gender_N)

# Display the values
Gender_N
```

### 🔹 Step 3: Plot the Data

```r
# Plot response variable against Gender
plot(Gender_N ~ Gender,
     ylim = c(1.75, 2.00),
     main = "Fixed Effect Model: Gender Comparison",
     xlab = "Gender",
     ylab = "Expected Value")
```

This plot shows the two group means (1.85 for Male and 1.89 for Female).

### 🔹 Step 4: Combine into a Data Frame

```r
# Create data frame
Data <- data.frame(Gender, Gender_N)

# View dataset
Data
```

## 📌 Statistical Model – Adding Residuals (Within-Group Variance)

To turn the deterministic model into a statistical model, we add:

* Random residuals (within-group variation)
* Assume both groups share the same variance

* Model notation: y∼x

(where Length depends on Gender)

### 🔹 Step 5: Add Residuals to the Expected Values

```r
# Set seed for reproducibility
set.seed(956)

# Generate residuals:
# 30 observations from N(0, 0.05)
Residuals <- rnorm(n = 30, mean = 0, sd = 0.05)

# Create response variable (Length)
Length <- Gender_N + Residuals
```

### 🔹 Step 6: Plot the Simulated Data

```r
# Basic plot
plot(Length ~ Gender,
     main = "Statistical Model with Residual Variation",
     xlab = "Gender",
     ylab = "Length",
     ylim = c(1.75, 2.00))

# Add points for clarity
points(Length ~ Gender)
```

__👉 This now resembles a real experiment because variability is present.__

## 📌 Alternative Approach: Direct Simulation Using rnorm()

Instead of manually adding residuals, we can directly simulate using group means.

### 🔹 Step 7: Simulate Directly from Group Means

```r
# Set seed again for reproducibility
set.seed(956)

# Generate data directly using group-specific means
Length2 <- rnorm(n = length(Gender_N),
                 mean = Gender_N,
                 sd = 0.05)

# Create data frame
data2 <- data.frame(Length2, Gender)

# View dataset
data2
```

### 🔹 Step 8: Plot Alternative Simulation

```r
plot(Length2 ~ Gender,
     main = "Direct Simulation using rnorm()",
     xlab = "Gender",
     ylab = "Length",
     ylim = c(1.75, 2.00))

points(Length2 ~ Gender)
```

### 🔹 Step 9: Perform a Two-Sample t-test

Now we test whether the group means differ.

__To Test:__

__Null Hypothesis (H₀)__: Mean(Male) = Mean(Female)

__Alternative Hypothesis (H₁)__: Means are different

```r
t.test(Length ~ Gender)
```

The t-test should detect the mean difference (1.85 vs 1.89) if the effect size is large relative to the SD (0.05).
