## 📘 Simulation of Statistical Models - One-Way ANOVA – Fixed Effect Model


### 📌 One-Way ANOVA – Fixed Effect Model
The One-Way ANOVA model is:

$Y_{ij}= 𝜇 + 𝜏_i + 𝜀_{ij}$

where

* 𝜇 = overall mean

* $𝜏_i$ = treatment effect

* $𝜀_{ij} ∼ N(0, σ^2)$

We extend the factor from 2 levels to 4 levels, called Treat.

🔹 __Step 1: Define the Factor (Independent Variable)__

```r
# Create a factor with 4 treatment levels
# 5 replications per level (total = 20 observations)

Treat <- gl(n = 4, k = 5,
            labels = c("Treat1", "Treat2", "Treat3", "Treat4"))

# Check structure
str(Treat)

# View factor
Treat
```

🔹__Step 2: Define the Expected Group Means__

```r
# Assign expected means to each treatment
Treat_N <- rep(c(2.0, 2.3, 2.1, 2.5), each = 5)

# Combine into a data frame to verify alignment
data <- data.frame(Treat, Treat_N)

data
```

🔹 __Step 3: Check Data Structure and Plot Expected Values__

```r
# Check structure of numerical vector
str(Treat_N)

# Plot expected values
plot(Treat_N ~ Treat,
     main = "Expected Treatment Means",
     xlab = "Treatment",
     ylab = "Expected Response",
     ylim = c(1.8, 2.7))
```

📌 __Step 4: Add Residuals (Within-Group Variation)__

Since we have 20 observations, we generate 20 residuals.

```r
# Set seed for reproducibility
set.seed(354)

# Generate residuals from N(0, 0.06)
Residuals <- rnorm(n = length(Treat),
                   mean = 0,
                   sd = 0.06)

# Create response variable
Response <- Treat_N + Residuals
```

🔹 __Step 5: Plot Simulated Data__

```r
plot(Response ~ Treat,
     main = "Simulated One-Way ANOVA Data",
     xlab = "Treatment",
     ylab = "Response",
     ylim = c(1.8, 2.7))

points(Response ~ Treat)
```

Now the data resemble a real experiment with variability.

📌 __Step 6: Perform One-Way ANOVA Using lm()__

🔹 __(A) Model Without Intercept (-1)__

This estimates absolute group means directly.

```r
Model1 <- lm(Response ~ Treat - 1)

summary(Model1)
```

👉 -1 removes the intercept, so each coefficient represents the mean of each treatment group.

🔹 __(B) Standard ANOVA Model (With Intercept)__

```r
Model2 <- lm(Response ~ Treat)

summary(Model2)
```

👉 Here:

* Intercept = mean of reference group (Treat1 by default)
* Other coefficients = difference from Treat1
