## 📘 Simulation of Statistical Models - Linear Regression (Simple Linear Model)

### 📌Linear Regression (Simple Linear Model):
We consider two types of numerical explanatory variables:

1. Randomly generated (uncontrolled) concentrations
2. Controlled laboratory concentrations

🔹 __Case 1: Randomly Generated Explanatory Variable__

__Example:__ Cd concentration in forest streams (values between 0 and 5).

Step 1: Generate Cd Concentrations (Uniform Distribution)

```r
# Set seed for reproducibility
set.seed(243)

# Generate 18 concentrations between 0 and 5
# round() gives realistic measurement precision
Conc_Cd <- round(runif(n = 18, min = 0, max = 5), 2)

# Check structure
str(Conc_Cd)
```

⚠ If using a normal distribution, be cautious about generating negative values.

🔹 __Case 2: Controlled Numerical Explanatory Variable__

In laboratory experiments, concentrations are fixed and replicated.

```r
# Set seed
set.seed(762)

# Controlled concentration levels (0 to 5), each replicated 3 times
Conc_Cd2 <- rep(c(0, 1, 2, 3, 4, 5), each = 3)

# Check structure
str(Conc_Cd2)
```

### 📌 Building the Linear Regression Model

We define:

* Intercept (β₀) = 1.2
* Slope (β₁) = 3
* Residual standard deviation = 0.6

Model form: $Y = β_0 ​+ β_1​X + ε$

### 🔹 Model for Case 1 (Uncontrolled Concentrations)

__Step 1: Generate Response Variable__

```r
# Set seed
set.seed(635)

# Define parameters
Intercept <- 1.2
Slope <- 3

# Generate residuals
Residuals <- rnorm(n = length(Conc_Cd),
                   mean = 0,
                   sd = 0.6)

# Generate response variable
Diverse <- Intercept + Slope * Conc_Cd + Residuals
```

🔹__Step 2: Plot Data and Regression Line__

```r
plot(Diverse ~ Conc_Cd,
     main = "Linear Regression (Uncontrolled Concentrations)",
     xlab = "Cd Concentration",
     ylab = "Diaspore Diversity",
     xlim = c(0, 5),
     ylim = c(0, 20))

abline(lm(Diverse ~ Conc_Cd))
```

### 🔹 Model for Case 2 (Controlled Concentrations)
__Step 1: Generate Response Variable__

```r
# Set seed
set.seed(635)

# Define parameters
Intercept <- 1.2
Slope <- 3

# Generate residuals
Residuals <- rnorm(n = length(Conc_Cd2),
                   mean = 0,
                   sd = 0.6)

# Generate response variable
Diverse2 <- Intercept + Slope * Conc_Cd2 + Residuals
```

__Step 2: Plot Data and Regression Line__

```r
plot(Diverse2 ~ Conc_Cd2,
     main = "Linear Regression (Controlled Concentrations)",
     xlab = "Cd Concentration",
     ylab = "Diaspore Diversity")

abline(lm(Diverse2 ~ Conc_Cd2))
```

### 📌 Alternative Coding Approach

Instead of manually adding residuals:

```r
set.seed(635)

# Deterministic linear predictor
Model <- Intercept + Slope * Conc_Cd2

# Generate response directly from normal distribution
Diverse3 <- rnorm(n = length(Conc_Cd2),
                  mean = Model,
                  sd = 0.6)

# Compare results
data.frame(Diverse2, Diverse3)
```

✅ This produces the same result.

### 📌 Perform Regression Analysis

```r
Model_fit <- lm(Diverse ~ Conc_Cd)

summary(Model_fit)
```

### 📊 Interpretation of Output

From summary(Model_fit):

* Intercept estimate ≈ 1.2
* Slope estimate ≈ 3
* Residual standard error ≈ 0.6
(Example output: Residual standard error: 0.54)

This confirms that:

* The model successfully recovers the true parameters.
* The estimated residual standard error approximates the simulated SD (0.6).

#### ✅ Statistical Model:

$Y_i​ = 1.2 + 3X_i​ + ε_i​$

where:  $ε_i​ ∼ N(0, 0.6^2)$
